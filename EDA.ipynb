```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_prices = pd.read_csv(r'C:\Users\muham\OneDrive\Projects\House prices and economics\cleaned data (csv)\House_price_csv_mode.csv')
unemployment_rate = pd.read_csv('https://raw.githubusercontent.com/HAZCHEM234/NZ_House_labor_market_data/main/Clean%20data/Unemployment_index.csv')
immigration_rate = pd.read_csv(r'C:\Users\muham\OneDrive\Projects\House prices and economics\cleaned data (csv)\2018-2024 Immigration .csv')
rental_market = pd.read_csv('https://raw.githubusercontent.com/HAZCHEM234/NZ_House_labor_market_data/main/Clean%20data/rental_index.csv')
```

### House plot

```python
# Convert the 'Date' column to datetime format
house_prices['Date'] = pd.to_datetime(house_prices['Date'], format='%d/%m/%Y')

# Filter data from 2010 onwards
filtered_data = house_prices[house_prices['Date'] >= '2018-01-01']

# Plotting
plt.figure(figsize=(10, 6))

# Plot each column as a line graph
plt.plot(filtered_data['Date'], filtered_data['House Sale'], label='House Sale')
# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('House Price Analysis (2010 - Present)')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(filtered_data['Date'], filtered_data['Total value of housing stock '], label='Housing Stock Value')
plt.plot(filtered_data['Date'], filtered_data['Residential investment (GDP)'], label='Residential Investment')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('House Price Analysis (2018 - Present)')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

### Unemployment rate analysis

```python
# Convert the 'Date' column to datetime format
unemployment_rate['Date'] = pd.to_datetime(unemployment_rate['Date'], format='%d/%m/%Y')

# Filter data from 2015 onwards
filtered_data = unemployment_rate[unemployment_rate['Date'] >= '2018-01-01']

# Plotting
plt.figure(figsize=(10, 6))

# Plot unemployment rate as a line graph
plt.plot(filtered_data['Date'], filtered_data['Unemployment rate'], label='Unemployment Rate', color='b')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate (2015 - Present)')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
